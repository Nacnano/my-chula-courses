## Problem 1 : Integer program


Solve the following program by using linprog function and branch and bound method:
$$Objective : max(3x + 4y) $$
\begin{equation*}
s.t.
\begin{split}
  x + 2y & \leq 7 \\
  3x  - y & \geq 0 \\
  x -  y & \leq  2 \\
  x, y  \in Z^+ \cup \{0\} \\
\end{split}
\end{equation*}




```
import numpy as np
from scipy.optimize import linprog

# Define the coefficients of the objective function (for maximization)
# We need to minimize in linprog, so we multiply by -1
c = [-3, -4]  # Objective: max(3x + 4y)

# Define the inequality constraints (A_ub * [x, y] <= b_ub)
A = [
    [1, 2],   # x + 2y <= 7
    [-3, 1],  # 3x - y >= 0  -> -3x + y <= 0
    [1, -1]   # x - y <= 2
]
b = [7, 0, 2]

# Bounds for x and y (both non-negative and integer)
x_bounds = (0, None)
y_bounds = (0, None)

# Step 1: Solve the relaxed LP problem
res = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')

# Check if the solution is integer
def is_integer_solution(sol):
    return np.allclose(sol, np.round(sol))

# Print the results of the relaxed problem
if res.success:
    print(f"Relaxed LP solution: {res.x}")
    if is_integer_solution(res.x):
        print(f"Optimal integer solution found: {np.round(res.x)}, Objective value: {-res.fun}")
    else:
        print("Relaxed solution is not an integer. Proceeding to branch and bound.")
else:
    print("No solution found for the relaxed LP problem.")

# Step 2: Implement Branch and Bound (recursive approach)
def branch_and_bound(c, A, b, bounds, best_sol=None, best_val=None):
    # Solve the relaxed LP
    res = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

    if not res.success:
        return best_sol, best_val  # No feasible solution in this branch

    x_val = res.x
    obj_val = -res.fun  # Remember, we negated the objective coefficients for maximization

    if is_integer_solution(x_val):
        # If this solution is better than the best one found so far, update the best solution
        if best_val is None or obj_val > best_val:
            best_sol = np.round(x_val)
            best_val = obj_val
            print(f"New best integer solution: {best_sol}, Objective value: {best_val}")
        return best_sol, best_val

    # If not an integer solution, we need to branch
    fractional_indices = np.where(np.abs(x_val - np.round(x_val)) > 1e-5)[0]
    if len(fractional_indices) == 0:
        return best_sol, best_val

    # Pick the first variable that is fractional
    idx = fractional_indices[0]
    floor_val = np.floor(x_val[idx])
    ceil_val = np.ceil(x_val[idx])

    # Branch 1: x[idx] <= floor_val
    new_bounds_1 = bounds.copy()
    new_bounds_1[idx] = (bounds[idx][0], floor_val)
    best_sol, best_val = branch_and_bound(c, A, b, new_bounds_1, best_sol, best_val)

    # Branch 2: x[idx] >= ceil_val
    new_bounds_2 = bounds.copy()
    new_bounds_2[idx] = (ceil_val, bounds[idx][1])
    best_sol, best_val = branch_and_bound(c, A, b, new_bounds_2, best_sol, best_val)

    return best_sol, best_val

# Initial bounds for both variables
initial_bounds = [(0, None), (0, None)]

# Step 3: Start the branch and bound process
best_solution, best_value = branch_and_bound(c, A, b, initial_bounds)

if best_solution is not None:
    print(f"Optimal integer solution: {best_solution}, Objective value: {best_value}")
else:
    print("No feasible integer solution found.")
```

```
import numpy as np # you could use pandas to convert the array into csv
town_location = np.array([[  57.1817396 , -100.21311863],
       [  27.20599853,  -83.10295469],
       [-151.77984602, -204.29167019],
       [  45.64179719,   46.87585513],
       [-169.69698481, -110.76447501],
       [  87.25935223,  -91.63706912],
       [ 144.11419822,  200.79810504],
       [ -63.81741004,   43.32842009],
       [  92.65947426,  -10.32326513],
      ])
```

```
def visualize_factory_coverage(town_location, factory_idx):
  '''
  visualize the factory and town locations.
  Input:
    town_location : location of town in (x, y) coordinate
    factory_idx : the id of towns chosen to plant a factory
  Output:
    visual output displaying town (red), factory (green) and factory's  delivery range(circle)
  '''
  factory_location = town_location[[factory_idx]]
  import matplotlib.pyplot as plt
  fig, ax = plt.subplots()
  plt.xlim(-300, 300)
  plt.ylim(-300, 300)
  plt.scatter(town_location[:,0], town_location[:,1], color = 'red')
  plt.scatter(factory_location[:,0], factory_location[:,1], color = 'green')
  ax.set_aspect(1)
  for x in factory_location:
    ax.add_artist(plt.Circle((x[0],x[1]), 50, fill=False, color = 'blue'))

  plt.show()
visualize_factory_coverage(town_location, [0, 2, 3])
```

## Problem 5: Quantile Regression (piecewise linear function)

If you observe the objective function from the problem 6 in the previous lab, the solution that minimizes $l_2$ regression (least square method) and $l_1$ regression is the mean and median of the dataset respectively.  Quantile regression is a generalized formed of $l_1$ regression you have solved in the previous question. Likewise, The solution of quantile regression is the $q^{th}$ quantile of the dataset. A mathematical program for quantile regression is:

####Decision variable $\beta_1, \beta_0$

$$Objective : min( \sum_{i=1}^{N}L_q(y_i - (\beta_1x_i + \beta_0))) $$
\begin{equation*}
s.t.
\begin{split}
  \beta_1, \beta_0 \in R
\end{split}
\end{equation*}

where
\begin{equation*}
\begin{split}
  L_q(x) = \left\{
                \begin{array}{ll}
                  (q-1)x, & x \leq 0\\
                  qx, & x > 0\\
                \end{array}
              \right.
\end{split}
\end{equation*}

Thus, a $l_1$ regression is equivalent to a quantile regression where q = 0.5.
